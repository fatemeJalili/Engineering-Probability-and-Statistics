```python
import numpy as np
import matplotlib.pyplot as plt

data_U = np.random.uniform(0, 1, 1000)
data_V = np.random.uniform(0, 1, 1000)
R = np.sqrt(-2 * np.log(data_V))

data_X = R * np.cos(2 * np.pi * data_U)
data_Y = R * np.sin(2 * np.pi * data_U)

plt.hist(data_X, bins = 50,color = 'lightblue')
plt.title('histogram of X')
plt.grid(color = 'royalblue', linestyle = '--', linewidth = 0.25)
plt.show()

plt.hist(data_Y, bins = 50,color = 'lightblue')
plt.title('histogram of Y')
plt.grid(color = 'royalblue', linestyle = '--', linewidth = 0.25)
plt.show()
```
