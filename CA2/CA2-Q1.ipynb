part A


```python
import numpy as np
import matplotlib.pyplot as plt
import math

mu_X = 10
var_X = 16
mu_Y = 15
var_Y = 25
sigma_X = math.sqrt(var_X)
sigma_Y = math.sqrt(var_Y)

data_X = np.random.normal(mu_X, sigma_X, 1000)
data_Y = np.random.normal(mu_Y, sigma_Y, 1000)
```

part B



```python

plt.hist(data_X, bins = 50,color = 'deepskyblue')
plt.title('histogram of X')
plt.grid(color = 'royalblue', linestyle = '--', linewidth = 0.25)
plt.show()

plt.hist(data_Y, bins = 50,color = 'deepskyblue')
plt.title('histogram of Y')
plt.grid(color = 'royalblue', linestyle = '--', linewidth = 0.25)
plt.show()
```

part C

```python
# zipped_lists` contains pairs of items from both lists.
# Create a list with the sum of each pair.
data_Z = [x + y for (x, y) in zip(data_X, data_Y)] 
```

part E

```python
import statistics

mu_Z = statistics.mean(data_Z)
var_Z = statistics.variance(data_Z)
print("mean of new random distributed variable Z is" ,mu_Z)
print("variance of new random distributed variable Z is" ,var_Z)
```
