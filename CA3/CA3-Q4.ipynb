```python
import random
import math
import numpy as np 
import matplotlib.pyplot as plt

def func1(x):
    return x**3
def func2(x):
    return np.sin(x)
def func3(x):
    return (1/(1+x**2))

def integral(a,b,function,N):
    int = 0
    data = np.zeros(N)
    for i in range(N):
        data[i] = random.uniform(a,b)
    for j in data:
        int += function(j)
    return int*(b-a)/float(N-1)
    
print('the first integral numerical value is ',integral(0, 1, func1,1000)) 
xs = range(2, 1000)
ys1 = [integral(0, 1, func1, x) for x in xs]
plt.plot(xs, ys1,':',label ="estimation",color = 'deepskyblue')
ys2 = [1/4 for x in xs]
plt.plot(xs, ys2,label = "exact value",color = 'darkviolet', linewidth = 1)
plt.title("Convergence diagram")
plt.xlabel('times of simulation')
plt.ylabel('estimation of fisrt integral')
plt.legend()
plt.show()

print('the second integral numerical value is ',integral(0,np.pi,func2,1000))
xs = range(2, 1000)
ys1 = [integral(0, np.pi, func2, x) for x in xs]
plt.plot(xs, ys1,':',label ="estimation",color = 'deepskyblue')
ys2 = [2 for x in xs]
plt.plot(xs, ys2,label = "exact value",color = 'darkviolet', linewidth = 1)
plt.title("Convergence diagram")
plt.xlabel('times of simulation')
plt.ylabel('estimation of second integral')
plt.legend()
plt.show()

print('the third integral numerical value is ',integral(0, 4, func3,1000))
xs = range(2, 1000)
ys1 = [integral(0, 4, func3, x) for x in xs]
plt.plot(xs, ys1,':',label ="estimation",color = 'deepskyblue')
ys2 = [math.atan(4) for x in xs]
plt.plot(xs, ys2,label = "exact value",color = 'darkviolet', linewidth = 1)
plt.title("Convergence diagram")
plt.xlabel('times of simulation')
plt.ylabel('estimation of third integral')
plt.legend()
plt.show()
```
