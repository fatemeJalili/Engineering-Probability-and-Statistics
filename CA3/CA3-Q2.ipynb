```python
import numpy as np
import math
import matplotlib.pyplot as plt

def estimate(numOfRounds):
    sum_n = 0
    for i in range(numOfRounds):
        sum = 0
        n = 0
        while sum <= 1:
            u = np.random.uniform(0, 1, 1)
            sum = sum + u
            n += 1
        sum_n = sum_n + n
    return(sum_n / numOfRounds)


#plot convergence diagram
xs = range(1, 1000)
ys1 = [estimate(x) for x in xs]
plt.plot(xs, ys1,':',label ="estimation",color = 'deepskyblue')
#also plot the exact form to compare
ys2 = [math.e for x in xs]
plt.plot(xs, ys2,label = "euler number",color = 'darkviolet', linewidth = 1)
#set title
plt.title("Convergence diagram")
#set label of x axis
plt.xlabel('times of simulation')
#set lable of y axis
plt.ylabel('estimation of e')
#plot two diagrams together
plt.legend()
#show plot
plt.show()
```
