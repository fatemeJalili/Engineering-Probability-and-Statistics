# method 1

```python
import numpy as np
from math import sqrt,cos,sin,pi
import matplotlib.pyplot as plt

theta1 = np.random.uniform(0, 2*pi, 1000)
theta2 = np.random.uniform(0, 2*pi, 1000)

X1 = np.cos(theta1)
Y1 = np.sin(theta1)
X2 = np.cos(theta2)
Y2 = np.sin(theta2)

distance = np.sqrt((X2-X1)**2+(Y2-Y1)**2)

figure, axes = plt.subplots()
circle = plt.Circle((0, 0), 1, color='black', fill=False, linewidth=5)
axes.set_aspect( 1 )
axes.add_artist( circle )
plt.plot((X1,X2),(Y1,Y2), linewidth=0.3)
plt.plot((sqrt(3)/2, 0, -sqrt(3)/2, sqrt(3)/2), (-0.5, 1, -0.5, -0.5), color='black', linewidth = 2)
plt.show

success = 0
for i in range(1000):
    if distance[i]> sqrt(3):
        success+=1
    
probability = success/1000
print("probability in fisrt method using 1000 times of simulation is :", probability)
```

# method 2

```python
import numpy as np
from math import sqrt,cos,sin,pi
import matplotlib.pyplot as plt

theta = np.random.uniform(0, 2*pi, 1000)
P = np.random.uniform(0, 1 , 1000)
Q = np.sqrt(1 - P**2)

X1 = P * np.cos(theta) + Q * np.sin(theta)
X2 = P * np.cos(theta) - Q * np.sin(theta)
Y1 = P * np.sin(theta) - Q * np.cos(theta)
Y2 = P * np.sin(theta) + Q * np.cos(theta)

distance = np.sqrt((X2-X1)**2+(Y2-Y1)**2)

figure, axes = plt.subplots()
circle = plt.Circle((0, 0), 1, color='black', fill=False, linewidth=5)
axes.set_aspect( 1 )
axes.add_artist( circle )
plt.plot((X1,X2),(Y1,Y2), linewidth=0.3)
plt.plot((sqrt(3)/2, 0, -sqrt(3)/2, sqrt(3)/2), (-0.5, 1, -0.5, -0.5), color='black', linewidth = 2)
plt.show

success = 0
for i in range(1000):
    if distance[i]> sqrt(3):
        success+=1
    
probability = success/1000
print("probability in second method using 1000 times of simulation is :", probability)
```
