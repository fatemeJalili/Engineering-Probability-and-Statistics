```python
import random
import numpy as np
import matplotlib.pyplot as plt

def simulate(N):
    datak = [0]*1000
    for i in range(1000):
        num0 = 0
        num1 = 0
        while num0 < N+1 and num1 < N+1 :
            if random.randint(0,1) : num1 += 1
            else : num0 += 1
        if num0 == N+1: 
            datak[i] = N-num1
        else : 
            datak[i] = N-num0
    return datak
```

```python
plt.hist(simulate(20), bins = 50,color = 'lightblue')
plt.title('histogram of left mathes, N=20')
plt.grid(color = 'royalblue', linestyle = '--', linewidth = 0.25)
plt.show()

plt.hist(simulate(40), bins = 50,color = 'lightblue')
plt.title('histogram of left mathes, N=40')
plt.grid(color = 'royalblue', linestyle = '--', linewidth = 0.25)
plt.show()

plt.hist(simulate(60), bins = 50,color = 'lightblue')
plt.title('histogram of left mathes, N=60')
plt.grid(color = 'royalblue', linestyle = '--', linewidth = 0.25)
plt.show()


```
