# Same Birthday


part A



```python
from random import randint
from collections import Counter

#set times of simulation
timesOfSIM = 10000
#set a variable to count times of successful simulations
only2sameBday = 0

for i in range(timesOfSIM):
    #generate 23 random numbers between 1 to 365 and count repeated time of each one
    sameBdayRep = Counter([randint(1, 365) for person in range(23)])
    #set a list and save only repeated times
    bdayRep_sequence = sameBdayRep.values()
    #sort repeated times
    sortedBdaysRep = sorted(bdayRep_sequence)
    #check if we have only one random number as birhday date which has been repeated only two times
    if sortedBdaysRep[-2:] == [1, 2]:
        only2sameBday += 1
#calculate pbty
pbty_23 = (only2sameBday)/(timesOfSIM)

print("probability of exactly two same birthday between 23 person in 365 days with 10000 times of simulation is:",pbty_23)
```

part B

```python
def pbtySameBday(n,k):
    from random import randint
    from collections import Counter

    #set times of simulation
    timesOfSIM = 10000
    #set a variable to count times of successful simulations
    only2sameBday = 0

    for i in range(timesOfSIM):
        #generate k random numbers between 1 to n and count repeated time of each one
        sameBdayRep = Counter([randint(1, n) for person in range(k)])
        #set a list and save only repeated times
        bdayRep_sequence = sameBdayRep.values()
        #sort repeated times
        sortedBdaysRep = sorted(bdayRep_sequence)
        #check if we have only one random number as birhday date which has been repeated only two times
        if sortedBdaysRep[-2:] == [1, 2]:
            only2sameBday += 1
    #calculate pbty
    pbty = (only2sameBday)/(timesOfSIM)
    
    return(pbty)
    
```

part c

```python
#plot diagram
import matplotlib.pyplot as plt
xs = range(1, 81)
ys = [pbtySameBday(365,k) for k in xs]
plt.plot(xs, ys, linewidth=2.5, color = 'darkorchid')
#set plot title
plt.title("part C diagram")
#set label of x axis
plt.xlabel("k")
#set label of y axis
plt.ylabel('probability of 2 same Bday in a year')
#set digram grids
plt.grid(color = 'royalblue', linestyle = '--', linewidth = 0.25)
#show plot
plt.show()
```
